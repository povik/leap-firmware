```python
# plot the transfer function of a digital IIR filter derived
# from n-th order Butterworth, for couple of values of n

x = np.arange(0, 3.14, 0.001)
plt.ylim(-81, 1)
plt.grid()
w_c = 0.5 # corner angular frequency
for n in range(1, 6):
    f = butter(n)(Rational([-1.0, 1.0], [1.0, 1.0]) / math.tan(w_c / 2))
    plt.plot(x, np.log10(np.abs(f(np.exp(1j*x)))**2)*10, label=f"{n}")
plt.legend()
```
